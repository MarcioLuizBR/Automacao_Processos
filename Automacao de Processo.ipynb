```python
import pandas as pd
import pathlib
import smtplib
import email.message
import mimetypes
import unicodedata

from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders

import senhasegura
```

```python
# importar base de dados

emails = pd.read_excel(r"Bases de Dados\Emails.xlsx")
lojas = pd.read_csv(r"Bases de Dados\Lojas.csv", sep=";", encoding='latin1')
vendas = pd.read_excel(r"Bases de Dados\Vendas.xlsx")

display(emails)
display(lojas)
display(vendas)
```

### Passo 2 - Definir Criar uma Tabela para cada Loja e Definir o dia do Indicador

```python
# incluir nome da loja em vendas
vendas = vendas.merge(lojas, on='ID Loja')

display(vendas)
```

```python
dicionario_lojas = {}
for loja in lojas['Loja']:
    dicionario_lojas[loja] = vendas.loc[vendas['Loja']==loja, :]

display(dicionario_lojas['Shopping Recife'])
display(dicionario_lojas['Bourbon Shopping SP'])
```

```python
# encontrar o dia indicador

dia_indicador = vendas['Data'].max()
print(dia_indicador)
```

### Passo 3 - Salvar a planilha na pasta de backup

```python
# Função para remover acentos
def remover_acentos(texto):
    return ''.join(
        c for c in unicodedata.normalize('NFD', texto)
        if unicodedata.category(c) != 'Mn'
    )


# identific ar se a pasta já existe
caminho_backup = pathlib.Path(r'Backup Arquivos Lojas')

arquivos_pasta_backup = caminho_backup.iterdir()

# escrevendo em List comprehension - para praticar
lista_nomes_backup = [arquivo.name for arquivo in arquivos_pasta_backup]

for loja in dicionario_lojas:
    # normalizar nome da loja
    loja_normalizada = remover_acentos(loja)
    
    if loja_normalizada not in lista_nomes_backup:
        nova_pasta = caminho_backup / loja_normalizada
        nova_pasta.mkdir(parents=True, exist_ok=True)  # Garante que a pasta seja criada
    
    # salvar dentro da pasta
    nome_arquivo = '{}_{}_{}.xlsx'.format(dia_indicador.month, dia_indicador.day, loja_normalizada)
    local_arquivo = caminho_backup / loja_normalizada / nome_arquivo
    
    dicionario_lojas[loja].to_excel(local_arquivo)
        
```

### Passo 4 - Calcular o indicador

```python
# definição de metas

meta_faturamento_dia = 1000
meta_faturamento_ano = 1650000
meta_qtdeprodutos_dia = 4
meta_qtdeprodutos_ano = 120
meta_ticketmedio_dia = 500
meta_ticketmedio_ano = 500
```

### Passo 5 - Enviar por e-mail para o gerente

```python
# enviar o email
def enviar_email(loja, faturamento_dia, faturamento_ano, qtde_produtos_dia, qtde_produtos_ano, ticket_medio_dia, ticket_medio_ano, dia_indicador):  
    nome = emails.loc[emails['Loja'] == loja, 'Gerente'].values[0]
    email_gerentes = emails.loc[emails['Loja'] == loja, 'E-mail'].values[0]
    
    cor_fat_dia = 'green' if faturamento_dia >= meta_faturamento_dia else 'red'
    cor_fat_ano = 'green' if faturamento_ano >= meta_faturamento_ano else 'red'
    cor_qtde_dia = 'green' if qtde_produtos_dia >= meta_qtdeprodutos_dia else 'red'
    cor_qtde_ano = 'green' if qtde_produtos_ano >= meta_qtdeprodutos_ano else 'red'
    cor_ticket_dia = 'green' if ticket_medio_dia >= meta_ticketmedio_dia else 'red'
    cor_ticket_ano = 'green' if ticket_medio_ano >= meta_ticketmedio_ano else 'red'
    
    corpo_email = f"""
    <p>Bom dia {nome}</p>
    
    <p>O resultado de ontem <strong>({dia_indicador.day}/{dia_indicador.month})</strong> da <strong>Loja {loja}</strong></p>
    
    <table>
        <tr>
            <th>Indicador</th>
            <th>Valor dia</th>
            <th>Meta dia</th>
            <th>Cenário Dia</th>
        </tr>
        <tr>
            <td>Faturamento</td>
            <td style="text-align: center">R${faturamento_dia:.2f}</td>
            <td style="text-align: center">R${meta_faturamento_dia:.2f}</td>
            <td style="text-align: center"><font color="{cor_fat_dia}">◙</font></td>
        </tr>
        <tr>
            <td>Diversidade de Produtos</td>
            <td style="text-align: center">{qtde_produtos_dia}</td>
            <td style="text-align: center">{meta_qtdeprodutos_dia}</td>
            <td style="text-align: center"><font color="{cor_qtde_dia}">◙</font></td>
        </tr>
        <tr>
            <td>Ticket Médio</td>
            <td style="text-align: center">R${ticket_medio_dia:.2f}</td>
            <td style="text-align: center">R${meta_ticketmedio_dia:.2f}</td>
            <td style="text-align: center"><font color="{cor_ticket_dia}">◙</font></td>
        </tr>
    </table>
    <br>
    <table>
        <tr>
            <th>Indicador</th>
            <th>Valor Ano</th>
            <th>Meta Ano</th>
            <th>Cenário Ano</th>
        </tr>
        <tr>
            <td>Faturamento</td>
            <td style="text-align: center">R${faturamento_ano:.2f}</td>
            <td style="text-align: center">R${meta_faturamento_ano:.2f}</td>
            <td style="text-align: center"><font color="{cor_fat_ano}">◙</font></td>
        </tr>
        <tr>
            <td>Diversidade de Produtos</td>
            <td style="text-align: center">{qtde_produtos_ano}</td>
            <td style="text-align: center">{meta_qtdeprodutos_ano}</td>
            <td style="text-align: center"><font color="{cor_qtde_ano}">◙</font></td>
        </tr>
        <tr>
            <td>Ticket Médio</td>
            <td style="text-align: center">R${ticket_medio_ano:.2f}</td>
            <td style="text-align: center">R${meta_ticketmedio_ano:.2f}</td>
            <td style="text-align: center"><font color="{cor_ticket_ano}">◙</font></td>
        </tr>
    </table>
    
    <p> Segue em anexo a planilha com todos os dados para mais detalhes </p>
    
    <p> Qualquer dúvida estou à disposição. </p>
    <p> Att., Marcio Luiz </p>
    """
    msg = MIMEMultipart()
    msg['Subject'] = "OnePage Dia {}/{} - Loja {}".format(dia_indicador.day, dia_indicador.month, loja)
    msg['From'] = "marcio.asriel@gmail.com"
    msg['To'] = email_gerentes
    password = senhasegura.senha
        

    # Adicionando o corpo do e-mail
    msg.attach(MIMEText(corpo_email, 'html'))
    
    # Normalizar o nome da loja
    loja_normalizada = remover_acentos(loja)

    # Anexando um arquivo
    anexo_caminho = pathlib.Path.cwd() / caminho_backup / loja_normalizada / f'{dia_indicador.month}_{dia_indicador.day}_{loja_normalizada}.xlsx' # substitua pelo caminho do seu arquivo
   
        

    attachment = MIMEBase('application', 'octet-stream')
    try:
        with open(anexo_caminho, 'rb') as anexo:
            attachment.set_payload(anexo.read())
        encoders.encode_base64(attachment)
        attachment.add_header('Content-Disposition', f'attachment; filename="{anexo_caminho.name}"')
        msg.attach(attachment)
    except FileNotFoundError:
        print(f"Arquivo {anexo_caminho} não encontrado.")

    # Enviando o e-mail
    s = smtplib.SMTP('smtp.gmail.com: 587')
    s.starttls()
    s.login(msg['From'], password)
    s.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))
    s.quit()
    



for loja in dicionario_lojas:
    
    vendas_loja = dicionario_lojas[loja]
    vendas_loja_dia = vendas_loja.loc[vendas_loja['Data']==dia_indicador, :]


    # faturamento
    faturamento_ano = vendas_loja['Valor Final'].sum()
    faturamento_dia = vendas_loja_dia['Valor Final'].sum()
    
    # diversidade de produtos
    qtde_produtos_ano = len(vendas_loja['Produto'].unique())
    qtde_produtos_dia = len(vendas_loja_dia['Produto'].unique())

    # ticket medio
    valor_venda = vendas_loja.groupby('Código Venda').sum(numeric_only=True)
    ticket_medio_ano = valor_venda['Valor Final'].mean()

    # ticket medio dia
    valor_venda_dia = vendas_loja_dia.groupby('Código Venda').sum(numeric_only=True)
    ticket_medio_dia = valor_venda_dia['Valor Final'].mean()
    

    enviar_email(loja, faturamento_dia, faturamento_ano, qtde_produtos_dia, qtde_produtos_ano, ticket_medio_dia, ticket_medio_ano, dia_indicador)
    print(f'Email enviado para a loja {loja}')
    
    
```

### Passo 7 - Criar ranking para diretoria

### Passo 8 - Enviar e-mail para diretoria

```python
# enviar o email
def enviar_email_diretoria():  
    # Verificando se a loja "Diretoria" está presente
    if 'Diretoria' not in emails['Loja'].values:
        print("A loja 'Diretoria' não foi encontrada no DataFrame de e-mails.")
        return
    
    email_diretoria = emails.loc[emails['Loja'] == 'Diretoria', 'E-mail'].values

    # Verificando se o e-mail foi encontrado
    if not email_diretoria.size:  # Verifica se o array é vazio
        print("E-mail da diretoria não encontrado.")
        return  # Saia da função se não encontrar o e-mail
    
    email_diretoria = email_diretoria[0]  # Pegue o primeiro e-mail encontrado

    # Verificando se os DataFrames de faturamento estão preenchidos
    if faturamento_lojas_dia.empty:
        print("O DataFrame 'faturamento_lojas_dia' está vazio.")
        return
    if faturamento_lojas_ano.empty:
        print("O DataFrame 'faturamento_lojas_ano' está vazio.")
        return

    # Verificando se existem dados suficientes para acessar o índice
    if faturamento_lojas_dia.shape[0] == 0 or faturamento_lojas_ano.shape[0] == 0:
        print("Não há dados suficientes para determinar o melhor ou pior faturamento.")
        return

    corpo_email = f'''
    Prezados, bom dia
    
    Melhor loja do Dia em faturamento: Loja {faturamento_lojas_dia.index[0]} com o Faturamento R${faturamento_lojas_dia.iloc[0, 0]:.2f}
    Pior loja do Dia em Faturamento: Loja {faturamento_lojas_dia.index[-1]} com o Faturamento R${faturamento_lojas_dia.iloc[-1, 0]:.2f}
    
    Melhor loja do Ano em faturamento: Loja {faturamento_lojas_ano.index[0]} com o Faturamento R${faturamento_lojas_ano.iloc[0, 0]:.2f}
    Pior loja do Ano em Faturamento: Loja {faturamento_lojas_ano.index[-1]} com o Faturamento R${faturamento_lojas_ano.iloc[-1, 0]:.2f}
    
    Qualquer dúvida estou à disposição
    Att., Marcio Luiz 
    '''

    msg = MIMEMultipart()
    msg['Subject'] = f"Ranking Dia {dia_indicador.day}/{dia_indicador.month}"
    msg['From'] = "marcio.asriel@gmail.com"
    msg['To'] = email_diretoria
    password = senhasegura.senha

    # Adicionando o corpo do e-mail
    msg.attach(MIMEText(corpo_email, 'plain'))  # Altere para 'plain' se não houver HTML

    # Anexando os arquivos
    for ranking in ['Ranking Anual', 'Ranking Dia']:
        anexo_caminho = pathlib.Path.cwd() / caminho_backup / f'{dia_indicador.month}_{dia_indicador.day}_{ranking}.xlsx'
        attachment = MIMEBase('application', 'octet-stream')
        try:
            with open(anexo_caminho, 'rb') as anexo:
                attachment.set_payload(anexo.read())
            encoders.encode_base64(attachment)
            attachment.add_header('Content-Disposition', f'attachment; filename="{anexo_caminho.name}"')
            msg.attach(attachment)
        except FileNotFoundError:
            print(f"Arquivo {anexo_caminho} não encontrado.")
    
    # Enviando o e-mail
    try:
        s = smtplib.SMTP('smtp.gmail.com: 587')
        s.starttls()
        s.login(msg['From'], password)
        s.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))
        s.quit()
        print('E-mail da Diretoria enviado')
    except Exception as e:
        print(f"Erro ao enviar e-mail: {e}")

enviar_email_diretoria()


    
```
